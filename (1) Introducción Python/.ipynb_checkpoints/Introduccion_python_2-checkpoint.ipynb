### Modo Edición/Comandos<a id="section_edicion_com"> </a>

Similar a otros editores, Jupyter incluye dos modos de trabajo:
1. Edición: Con este modo editamamos el contenido en una celda, la celda siendo editada se marca en color **Verde**
2. Conmando: Con este modo ejecutamos comando globales. Los comandos que afectan celdas particulares se ejecutarán sobre la celda marcada en **Azul**


**Modo Edición**

Este modo altera el contenido de las celdas, si la celda es de tipo código, todo lo que está en ella será interpretado como código de python siguiendo todo su sintaxis (comentarios con "#", reglas de identación, etc).

Para acceder al modo de edicion basta apretar "enter" en la celda deseada, o hacer click sobre ella.

```python
# escribe tu "Hello word"
```

**Modo Comandos** - Atajos

Para acceder al modo comando basta apretar "esc", con las flecha podemos navegar a la celda que queremos.

Este modo es un poco menos natural de usar dependiendo del IDE al que uno esta acostumbrado, pero existe pues tiene herramientas muy poderosas:
1. Agregar o borrar celdas: 
* Agregar: ["esc"] **a**(bove) / **b**(elow)
* Borrar: ["esc"] **dd**(elete)
2. Copiado/Cortar/Pegar de celdas
* Copiar: ["esc"] **c**(opy)
* Pegar: ["esc"] **v**(paste)
* Cortar: ["esc"] **x**(ut)
3. Buscar  y Reemplazar
* Buscar: ["esc"] **"f"**(ind). *A priori buscará  y reemplazara en la celdas seleccionadas*
4. Seleccionar Multiples
* Seleccionar celda siguiente: ["esc"] **"shift" + j** 
* Seleccionar celda previa: ["esc"] **"shift" + k** 
5. Separar o Juntar Celdas
* Juntar con la siguiente: ["esc"] **"shift" + m**(erge). *Si están seleccionadas multiples celdas, se unirán todas*
* Separar celda en el cursor (Modo Edicion) **"ctrl"+"shift" + -**(underscore)
6. Cambiar a modo Markdown/Codigo
* Atajo a Markdown: ["esc"] + **m**(markdown)
* Atajo a Codigo: ["esc"] + **y**

El ["esc"] hace referencia a que debemos estar en modo comando para que funcione la instrucción.

# Data types en Python

### Listas <a id="section_listas"> </a>

Una lista es una colección de datos ordenada e indexada de valores que son intercambiables. Dicha lista permite valores duplicados. Nos sirven mucho para almacenar nombres de variables que queremos filtrar en un dataframe entre otras cosas.

Veamos algunos ejemplos y cómo acceder a los valores de la lista:

**1) Creación de lista y cómo acceder a sus valores.**

```python
simple_list = ['UBA',2,3,4,'CCA',2.71] #Creamos la lista
print(simple_list)
print("="*30)

print("El primer valor de la lista es: ")
print(simple_list[0]) #Nos va a impirmir el primer valor. Tener en cuenta que Python siempre comienza desde el número 0 y no del 1

print("="*30)
print("El cuarto valor de la lista es: ")
print(simple_list[3]) #Nos va a trear el cuarto valor (recordar que comienza en 0)

print("="*30)
print("El último valor de la lista es: ")
print(simple_list[-1])#Nos va a traer el último valor de la lista

print("="*30)
print("El ante ultimo valor de la lista es: ")
print(simple_list[-2])#Nos va a traer el ante último valor de la lista

print("="*30)
print("La cantidad de valores de la lista es: ")
print(len(simple_list))#Nos va a imprimir el "largo" o la cantidad de valores que hay en la lista.
```

Además de mostrar que es una lista, en dicho código podemos ver como se accede a las valores de la lista.

**2) Cambiar entradas de una lista.**

```python
simple_list[3] = 52 #Vamos a modificar el cuarto valor de la lista por 52
print(simple_list[3]) # Recordar que el indice arranca en 0!

print("="*30)
print("La lista appendeando el valor 'Final' queda de la siguiente manera: ")
#Agregamos un valor a la lista, se appendea al final.
simple_list.append('Final') 
print(simple_list)

print("="*30)
print("Observemos ahora la longitud de la lista o la cantidad de variables: ")
print(len(simple_list))#Vemos que la lista es mas
```

**3) Copiar una lista - IMPORTANTE ya que esto rige también cuando queramos copiar un DataFrame**

Siempre que se quiera copiar una lista en una nueva variable, debemos utilizar el metodo .copy()

```python
simple_list_2 = simple_list.copy() #Se debe hacer de este modo
```

Ahora, simple_list_2 es independiente de simple_list, y las modificaciones que hagamos sobre la segunda lista no van a ser afectadas en la primera. En cambio, si nosotros haciamos simple_list_2 = simple_list, los cambios que realicemos en la segunda lista van a afectar a la original, cosa que **NO** queremos!

**4) Slicing de listas.**

Podemos acceder a multiples valores de la lista, realizando slicing (muy similar a R).

```python
nueva_list = [1,5,2,7,3,66,1923,11] #Creamos nueva lista
nueva_list
```

```python
print(nueva_list[2:5])
```

**Tener en cuenta que el lo que se encuentra delante de ":" es inclusivo, y lo que se encuentra a su derecha es exclusivo**.

Entonces, dicho filtro va a ir desde la 3ra posición hasata la quinta. Recordar siempre que python arranca desde 0!

```python
print(nueva_list[2:3]) #Esto nos va a traer un solo valor ya que la parte derecha es exclusiva.
```

Tambien podemos utilizar valores negativos:

```python
print(nueva_list[-5:-1])
```

Vemos que aquí nos va a traer valores desde la posición 5 (contando de derecha a izquierda), hasta la última posicion (exclusive).

## Diccionarios <a id="section_diccionarios"> </a>

Un diccionario es una coleccion de valores que que tienen un "key" - llave, y un valor. Es análogo a la definición de un diccionario común y corriente, en donde uno tiene una palabra (en este caso puede ser un número, un string, lo que sea), y una definición (en este caso la definición va a ser un valor el cual también puede ser númerico o string). Al igual que las listas, sus valores pueden ser modificados.

**1) Creación de diccionario y acceso a sus valores**

```python
#Creamos un diccionario
simple_dict = {
    "marca": "Apple",
    "producto": "iPhone",
    "modelo": "X"
}
print(simple_dict)
print('Compre un',simple_dict['producto'],"modelo",simple_dict['modelo'],'de',simple_dict['marca'])
```

Vemos que las keys de este diccionario son "marca", "modelo" y "producto", y para acceder a ellas solo basta ingresar el nombre del diccionario, seguido de ["key"]. Por ejemplo simple_dict['producto'] hace referencia a el valor "iPhone".

**2) Cambiar valores de un diccionario**

```python
simple_dict["modelo"] = "11 pro"#Cambiamos el valor del modelo del diccionario y veamos cómo queda la frase.
print('Compre un',simple_dict['producto'],"modelo",simple_dict['modelo'],'de',simple_dict['marca'])
```

**3) Agregar nuevas entradas al diccionario**

```python

simple_dict['color'] = 'rojo'
print('Compre',simple_dict["color"],simple_dict['producto'],"modelo",simple_dict['modelo'],'de',simple_dict['marca'])
```

##### Metodos

Los metodos son funciones que pueden hacer calculos u operaciones en los atirbutos o en data que se use como input de la clase.

**Veamos un ejemplo de aplicación en sklearn**

La libreria Sklearn esta orientada a aplicaciones y procesos que se suelen hacer en un Workflow a la hora de realizar un modelo de Machine Learning. En el siguiente ejemplo vamos a agarrar una columna del dataset creada previamente, y estandarizarla utilizando la clase de sklean StandardScaler().

Puede que este ejemplo requiera de algunos conocimientos un poco más avanzados, pero no se preocupen si no entienden algo, ya que luego se darán estos temas con mayor profundidad en el curso.

**1)Importamos la clase y la instanciamos**

```python
from sklearn.preprocessing import StandardScaler #Importamos

scaler = StandardScaler() #Instanciamos, vemos que solo instanciamos la clase, sin ningun parametro dentro.
```

Antes veamos el dataset para ver las variables que vamos a estandarizar.

```python
df.head()
```

Vamos a estandarizar la primer columna.

```python
df_a_estandarizar = df[[df.columns.tolist()[0]]]

df_a_estandarizar.head()
```

**Ahora, aplicamos la clase que instanciamos (scaler) a este dataframe.** De nuevo, no se preocupen si no entienden el código.

```python
df_estandarizado = pd.DataFrame(scaler.fit_transform(df_a_estandarizar)#Usamos el metodo fit_transform de la clase.
                                , columns = df_a_estandarizar.columns,
                                index=df_a_estandarizar.index)

df_estandarizado.head()
```

Podemos ver que ahora nuestras variables fueron estandarizadas, usando simplemente la media y el desvío de cada columna. En este caso fue sólo una columna a modo de ejemplo.

**Ahora, ¿cómo hizo esto la clase?**

La instancia scaler, ahora vamos a ver que tiene nuevos atributos, los cuales los aprendio cuando hicimos .fit_transform sobre el dataset.

```python
display(scaler.mean_) #Chequeamos el atributo que aprendio la clase (la media de la columna sepal_lenght)
```

mean_ en este caso es un atributo de la clase scaler.

Ahora podemos ver que la clase al correr el metodo fit_transform, aprendio el atributo mean, que luego fue utilizado para escalar la columna. Si queremos chequear que esa sea la media, podemos hacerlo facilmente.

```python
df[df.columns[0]].mean()
```

Ésto fue aprendido por la clase usado el método fit. Nosotros hicimos fit_transform entonces este metodo ademas de aprender la media, transformo la columna directamente.

# Funciones

En esta última parte, veremos brevemente como realizar funciones en python.

**1)**

```python
# Definimos una funcion que tome un input, lo multiplique por si mismo, y le sume 7.
def x2p7(x):
    """    Aca es donde se escribe la documentacion en string 
    para que el usuario sepa lo que hace la funcion.
    Esta funcion devuelve x^2 + 7"""
    y = x*x
    z = y+7
    return z
# Tener en cuenta que luego del def. las lineas estan indentadas.
```

Pongamosla en práctica

```python
print('x^2 + 7 = ',x2p7(5))
print('x^2 + 7 = ',x2p7(1))
print('x^2 + 7 = ',x2p7(3))


#Si queremos ver la documentacion
print(x2p7.__doc__)
```
