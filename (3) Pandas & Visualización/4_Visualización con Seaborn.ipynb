## 📚 Librerias

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as pdr
```

## 📖  Data Sets 
---
Seaborn incluye datasets para usar.

```python
tips = sns.load_dataset("tips")
tips.head()

titanic = sns.load_dataset("titanic")
titanic.head()
```

## 📉   Tipos De gráficos
---

🔴 **Graficos Relacionados con Variables Discretas:**


- Box  (x discreta , y continua)
- Boxen (x discreta, y continua)
- Count (x discreta)


🔵  **Gráficos solo Vairbles Continuas:**

- Densidad
- Join plot
- Matriz plot

### 🪀 Gráficos Tipo Boxplot  
---

Los gráficos de variables discretasse crean con **sns.catplot(kind= "" )**
- Box
- Boxen

```python
sns.catplot(kind="box", x="day", y="total_bill", data=tips,  height=5)
plt.ylabel("Propina")
plt.xlabel("Día")
plt.title("Box-plot Propina Según el Día", fontdict={"fontsize":20});
```

```python
sns.catplot(kind="boxen", x="day", y="total_bill", data=tips,  height=5)
plt.ylabel("Propina")
plt.xlabel("Día")
plt.title("Box-plot Propina Según el Día", fontdict={"fontsize":20});
```

#### 🎮 El parámetro hue
---
Este parámetro permite incluir uan tercer variable.

**hue -> variable discreta**


```python
sns.catplot(kind="boxen", x="day", y="total_bill",hue="smoker", data=tips,  height=5)
plt.ylabel("Propina")
plt.xlabel("Día")
plt.title("Box-plot Propina Según el Día", fontdict={"fontsize":20});
```

### 📊 Gráficos de barra
---
Los gráficos de variables discretasse crean con **sns.catplot(kind= "" )**.

```python
titanic.head()
```

```python
sns.catplot(x="embarked", kind="count",  data=titanic);
```

```python
sns.catplot(x="sibsp", kind="count",  data=titanic);
```

```python
# hue agrega otra variable
sns.catplot(x="embarked", hue="survived", kind="count",  data=titanic);
```

```python
sns.catplot(x="sex", hue="survived", kind="count",  data=titanic);
```

### 🎫 Gráficos de Distribución  
---

Este tipo de gráficos son el histograma  y la función de densidad. 

```python
x = np.random.normal(size=500)
sns.distplot(x, bins=50)
```

### 🛺  Gráfico Datos Bivariados
---
Dos variables continuas, combinación de **gráfico de distribución** con **scatter plot**

```python
sns.jointplot(x="age", y="fare", data=titanic);
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde");
```

### 🧻  Gráfico de Matriz
---
Combina **Scatter Plot** con **histogramas**.

Solo **variables continuas**


```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(data=iris);
```

```python
# Podemos incluir un variable categórica con hue
sns.pairplot(iris, hue="species");
```
