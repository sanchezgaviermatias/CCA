### 游 Gr치ficos Tipo Boxplot  
---

Los gr치ficos de variables discretasse crean con **sns.catplot(kind= "" )**
- Box
- Boxen

```python
sns.catplot(kind="box", x="day", y="total_bill", data=tips,  height=5)
plt.ylabel("Propina")
plt.xlabel("D칤a")
plt.title("Box-plot Propina Seg칰n el D칤a", fontdict={"fontsize":20});
```

```python
sns.catplot(kind="boxen", x="day", y="total_bill", data=tips,  height=5)
plt.ylabel("Propina")
plt.xlabel("D칤a")
plt.title("Box-plot Propina Seg칰n el D칤a", fontdict={"fontsize":20});
```

#### 游꿡 El par치metro hue
---
Este par치metro permite incluir uan tercer variable.

**hue -> variable discreta**


```python
sns.catplot(kind="boxen", x="day", y="total_bill",hue="smoker", data=tips,  height=5)
plt.ylabel("Propina")
plt.xlabel("D칤a")
plt.title("Box-plot Propina Seg칰n el D칤a", fontdict={"fontsize":20});
```
